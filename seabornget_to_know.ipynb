```python
import seaborn as sns
import pandas as pd
```

```python
#getting some default datasets haru
sns.get_dataset_names()
```

```python
#loading any 4
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
#scatter plot
sns.scatterplot(x="tip", y="total_bill", data=tips, hue = "sex", size="size")
```

```python
#histogram plot showcasing for total blls
#kde = kernel density estimate 
sns.histplot(tips['total_bill'], bins =15, kde =True)
```

```python
#barplot to see which sex have how mauch tips
sns.barplot(tips, x='sex', y='tip')
```

```python
#boxplot to show which day how much tips are earned with mid horizontal line being median and upper and lower horizontal
#lines being max and min with points being outliers
sns.boxplot(tips, x='day', y='tip', hue='sex')
```

```python
#stripplot
#dodge separates sex
sns.stripplot(tips, x='day', y='tip', hue='sex', dodge = True)
```

```python
#jointplot: basically scatter and distribution plot sangai
sns.jointplot(tips, x='tip',y='total_bill',kind='kde', fill=True)
```

```python
sns.jointplot(tips, x='tip',y='total_bill',kind='hex')
```

```python
sns.pairplot(tips.select_dtypes(['number']))
```

```python
tips.corr(numeric_only = True)
```

```python
sns.heatmap(tips.corr(numeric_only = True), annot=True, cmap='icefire')
```
